```python
resource_folders
```

```python
if not os.path.isdir(f"vectorstores/{company_name}"):
    os.mkdir(f"vectorstores/{company_name}")
```

```python
for folder in resource_folders:
    print(folder)
    if not os.path.isdir(f"vectorstores/{company_name}/{folder}"):
        os.mkdir(f"vectorstores/{company_name}/{folder}")
    
    files_specific=os.listdir(f"resources/{company_name}/{folder}")
#     print(files_specific)
    if ".DS_Store" in files_specific:
        files_specific.remove(".DS_Store")
    documents=[]
    for f in files_specific:
        print(f)
        if not f.endswith(".html"):
            continue
        html_path = f"resources/{company_name}/{folder}/{f}"
        print(html_path)
        loader = BSHTMLLoader(html_path)
        documents.extend(loader.load())
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
    texts=text_splitter.split_documents(documents)
    vectorstore = Chroma.from_documents(documents=texts, embedding=GPT4AllEmbeddings(),persist_directory=f"vectorstores/{company_name}/{folder}")      
    print("*"*8)
    
    # create the vector DB
    
```
